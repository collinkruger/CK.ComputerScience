```python
import numpy as np

# We're importing scipy because it happens
# to contain image manipulation functionality
# NOTE: scipy is more general purpose
#       than just image manipulation
from scipy import ndimage
from scipy import misc

# We're importing functionlality from matplotlib because
# it happens to contain image display functionality
# NOTE: matplotlib is more general purpose
#       than just image display
import matplotlib.pyplot as plt
```

```python
# Most commonly (always?), images are represented in numpy as
# 3-dimensional arrays. (height of image, width of image, number of channels)
# For instance if an image is a full HD image with RGB color, it's shape would be (1080, 1920, 3) (matrix conventions, not traditinal image conventions (ex: not 1920x1080x3)). 
# If the image was full HD but grayscale, it's shape would be (1080, 1920, 1).
# For color channels, each color is usually (always?) encoded as 0-255.
# For gray scale channels, each color is usualy (always?) encoded as 0.0-1.0.
```

```python
face = misc.face()
```

```python
# "imshow" takes an ndarray and returns an AxesImage
# Jupyter notebooks will render the chart
# WITHOUT returning it as the last value, though I am not sure why.
# See: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html
#      The documentation covers data formats it can render

display(plt.imshow(face))
plt.show()
```

```python
# In our case, since our image is encoded as an ndarray,
# we can perform common ndarray manipulations such as slicing.

print("slice")

sliced_ndarray = face[:384, :512, :]
plt.imshow(sliced_ndarray)
plt.show()
```

```python
# and splitting

print("\nvertical split")

vsplit_ndarray = np.vsplit(face, 2)
plt.imshow(vsplit_ndarray[0])
plt.show()


print("\nhorizontal split")

hsplit_ndarray = np.hsplit(face, 2)
plt.imshow(hsplit_ndarray[0])
plt.show()


print("\nsplit at arbitrary locations")

column_split_ndarray = np.split(face, [200, 824], axis=1)
for column in column_split_ndarray:
    plt.imshow(column)
    plt.show()
```

```python
# concatenating

print("reconstructed from vsplit")

plt.imshow(np.concatenate(vsplit_ndarray))
plt.show()


print("\nreconstructed from hsplit")

plt.imshow(np.concatenate(hsplit_ndarray, axis=1))
plt.show()
```

```python
# and manipulate it's numeric (pixel) values

plt.imshow((face / 2).astype(int))
plt.show()

plt.imshow(np.clip(face * 2, 0, 255))
plt.show()
```
